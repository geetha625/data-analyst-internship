```python
# create dataframe
import pandas as pd
data={
    "name":["ram","sam","tom"],
    "marks":[80,90,85]
}
df=pd.DataFrame(data)
```

```python
# filter students
df[df["marks"]>80]
```

```python
# replace missing values with 0
df.fillna(0)
```

```python
# find avg marks by name
df.groupby("name").mean()
```

```python
# find max marks
df["marks"].max()
```
