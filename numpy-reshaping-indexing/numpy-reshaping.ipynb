NUMPY RESHAPING ( changing structure )

reshaping=changing the shape of the array without changing the data
- same numbers,different layout

RULE : total elements must remain the same

```python
# reshape 1D to 2D
import numpy as np
arr=np.array([1,2,3,4,5,6])
new_arr=arr.reshape(2,3)    # 2=rows,3=columns
print(new_arr)
```

```python
import numpy as np
data=np.array([10,20,30,40,50,60])
reshaped=data.reshape(3,2)     # 3=rows,2=columns
print(reshaped)
print("element:",reshaped[1,0])   # 1=row(index) , 0=col(index)
```
