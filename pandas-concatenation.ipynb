PANDAS CONCATENATION AND INDEXING are used constantly when combining datasets or selecting data

PANDAS CONCATENATION = joining multiple data frames



```python
# CONCATENATE ROWS (axis=0)

import pandas as pd
df1=pd.DataFrame({
    "name":["A","B"],
    "age":[20,21]
})
df2=pd.DataFrame({
    "name":["C","D"],
    "age":[22,23]
})
result=pd.concat([df1,df2])
print(result)
```

```python
# CONCATENATE COLUMNS (axis=1)

result=pd.concat([df1,df2],axis=1)
print(result)
```

```python
# using ignore_index=True , we get index as sequence
result=pd.concat([df1,df2],ignore_index=True)
print(result)
```

PANDAS INDEXING

INDEXING = selecting data

there are 2 main methods 

1) .loc - label based indexing

syntax : df.loc[row_label,column_label]

```python
# example :
df=pd.DataFrame({
    "name":["A","B","C"],
    "marks":[80,90,85]
})
print(df.loc[1])
```

```python
# select specific value
df.loc[2,"marks"]
```

2. iloc - position based indexing

- df.iloc[row_position,col_position]

```python
# example :
df.iloc[0]
```

```python
# select specific value :
df.iloc[1,1]
```

PRACTICE

```python
# concatenate rows
import pandas as pd
df1=pd.DataFrame({
    "name":["A","B"],
    "marks":[80,90]
})
df2=pd.DataFrame({
    "name":["C"],
    "marks":[85]
})
pd.concat([df1,df2])
```

```python
# concatenate columns
import pandas as pd
df1=pd.DataFrame({
    "name":["A","B"]
})
df2=pd.DataFrame({
    "marks":[80,90]
})
pd.concat([df1,df2],axis=1)
```

```python
# reset index after concatenation
pd.concat([df1,df2],ignore_index=True)
```

```python
# using loc find age of sam
df=pd.DataFrame({
    "name":["ram","sam","tom"],
    "age":[21,22,23]
})
df.loc[1]
```

```python
# using iloc
df.iloc[2,1]
```

PANDAS DATAFRAMES SERIES

-SERIES : a series is a single column of data

```python
import pandas as pd
data=[10,20,30,40]
s=pd.Series(data)
print(s)
```

```python
# SERIES WITH CUSTOM INDEX
import pandas as pd
marks=[80,90,85]
s=pd.Series(marks,index=["ram","sam","tom"])
print(s)
```
