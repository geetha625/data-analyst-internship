PANDAS CONCATENATION AND INDEXING are used constantly when combining datasets or selecting data

PANDAS CONCATENATION = joining multiple data frames



```python
# CONCATENATE ROWS (axis=0)

import pandas as pd
df1=pd.DataFrame({
    "name":["A","B"],
    "age":[20,21]
})
df2=pd.DataFrame({
    "name":["C","D"],
    "age":[22,23]
})
result=pd.concat([df1,df2])
print(result)
```

```python
# CONCATENATE COLUMNS (axis=1)

result=pd.concat([df1,df2],axis=1)
print(result)
```

```python
# using ignore_index=True , we get index as sequence
result=pd.concat([df1,df2],ignore_index=True)
print(result)
```

PANDAS INDEXING

INDEXING = selecting data

there are 2 main methods 

1) .loc - label based indexing

syntax : df.loc[row_label,column_label]

```python
# example :
df=pd.DataFrame({
    "name":["A","B","C"],
    "marks":[80,90,85]
})
print(df.loc[1])
```
