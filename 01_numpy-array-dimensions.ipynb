NumPy is a python library used to work with numerical data efficiently

numpy array : faster.more powerful list

numpy is used to perform fast mathematical operations on large datasets using arrays

- it is faster
- easy to calculate
- better for large datasets
- more powerful than normal python lists
 we can store data in arrays and perform operations like :
 - mean
 - max
 - sum
 - calculations


```python
import numpy as np
numbers=np.array([10,20,30,40])
print(numbers)
```

```python
"find the average"
print(numbers.mean())
```

```python
"find max and min"
print(numbers.max())
print(numbers.min())
```

```python
import numpy as np
marks=np.array([75,82,90,68,88])
print("average:",marks.mean())
print("highest:",marks.max())
```

ARRAY DIMENSIONS

dimension=no of directions data is arranged

1d : a line

2d : a table

3d : multiple tables stacked

we use dimensions when working with 
- excel tables
- customer data
- sales reports

1-D array : this is like a normal list (one row)

```python
import numpy as np
arr1=np.array([10,20,30,40])
print(arr1)
print(arr1.ndim)
```

2-D array : this is like a table (rows and columns)



```python
arr2=np.array([[1,2,3],[4,5,6]])
print(arr2)
print(arr2.ndim)
```

3-D array : this is like multiple tables stacked together

```python
arr3=np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(arr3)
print(arr3.ndim)
```

NOTE :

arr.ndim  # no of dim

arr.shape # rows & columns

arr.size # no of elements

```python
import numpy as np
data=np.array([
    [1,2],
    [3,4],
    [5,6]
    ])
print("dim:",data.ndim)
print("shape:",data.shape)
```
