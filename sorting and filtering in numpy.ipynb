SORTING IN NUMPY

SORT A 1D ARRAY


```python
import numpy as np
arr=np.array([5,2,9,1])
sorted_arr=np.sort(arr)
print(sorted_arr)
```

SORT ROWS (axis=1)

```python
arr=np.array([[3,1,2],
              [9,7,8]])
sorted_arr=np.sort(arr,axis=1)
print(sorted_arr)
```

SORT COLUMNS (axis=0)

```python
arr=np.array([[3,1],
              [2,4]])
sorted_arr=np.sort(arr,axis=0)
print(sorted_arr)
```

FILTERING IN NUMPY

- filtering : select elements based on condition

- used in data cleaning , data analysis , machine learning

```python
# filter nums greater than 10
arr=np.array([5,12,7,20])
result=arr[arr>10]
print(result)
```

```python
# filter even nums
arr=np.array([1,2,3,4,5,6])
result=arr[arr%2==0]
print(result)
```

PRACTICE

```python
#1 sort a 1d array
arr=np.array([8,3,6,1])
sorted_arr=np.sort(arr)
print(sorted_arr)
```

```python
#2 sort rows
arr=np.array([[4,2,3],
              [9,1,7]])
result=np.sort(arr,axis=1)
print(result)
```

AGGREGATION FUNCTIONS IN NUMPY

- aggregation : calculating a single value from many values

EXAMPLES

- sum : total
- mean : average
- max : largest
- min : smallest

```python
# SUM - total of elements
import numpy as np
arr=np.array([10,20,30])
print(np.sum(arr))
```

```python
# SUM WITH AXIS 
arr=np.array([[1,2],
              [3,4]])
print(np.sum(arr,axis=1))
```

```python
print(np.sum(arr,axis=0))
```

```python
# MEAN - average
arr=np.array([10,20,30])
print(np.mean(arr))
```

```python
# MAX - largest value
arr=np.array([5,12,7])
print(np.max(arr))
```

```python
# MIN - smallest value
arr=np.array([5,12,7])
print(np.min(arr))
```

```python
# ARGMAX - index of largest value
arr=np.array([5,12,7])
print(np.argmax(arr))
```

```python
# ARGMIN - index of smallest value
arr=np.array([5,12,7])
print(np.argmin(arr))
```
