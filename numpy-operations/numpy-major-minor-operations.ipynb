MAJOR OPERATIONS - row wise operations

Major Operations = operations across rows

- axis = 1 [row operation]

```python
import numpy as np
data=np.array([[5,6],
               [8,9]])
np.sum(data,axis=1)
```

MINOR OPERATIONS - column wise operations

Minor Operations = operations across columns

- axis=0 [column operation]

```python
np.sum(data,axis=0)
```

COMMON MAJOR AND MINOR OPERATIONS 

- SUM

np.sum(data,axis=0) # column(minor)

np.sum(data,axis=1) # row(major)

- MEAN 

np.mean(data,axis=0) 

np.mean(data,axis=1)

- MAXIMUM

np.max(data,axis=0)

np.max(data,axis=1)

- MINIMUM

np.min(data,axis=0)

np.min(data,axis=1)

- COUNT

np.count_nonzero(data,axis=0)

```python
import numpy as np
data=np.array([[101,5000,25],
               [102,6000,30],
               [103,5500,28],
               [104,7000,35]])
# columns : id|salary|age
```

```python
#1 find sum of each col
np.sum(data,axis=0)
```

```python
# 2 find sum of each row
np.sum(data,axis=1)
```

```python
# 3 find the avg salary
np.mean(data[:,1],axis=0)
```

```python
# 4 find max age
np.max(data[:,2],axis=0)
```

```python
# 5 count employees
(data.shape[0])
```

```python
# 6 find max salary
np.min(data[:,1],axis=0)
```

```python
# 7 find mean of age
np.mean(data[:,2],axis=0)
```

```python
# 8 find the row with highest salary     [ to find row use arg]
np.argmax(data[:,1],axis=0)
```

```python
# 9 find the total salary of all employee
np.sum(data[:,1],axis=0)
```

```python
# 10 find column wise max values
np.max(data,axis=0)
```
