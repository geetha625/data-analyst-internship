```python
import pandas as pd
df=pd.read_csv("data_csv")            # used to read a file
```

read_csv() - it reads a csv file(comma separated value) & converts it into dataframe

df.head() - shows first 5 rows

df.tail() - shows last 5 rows

df.shape() - returns (row,col)

2. SELECTING DATA 

df["column"] - selects 1 column

df[["col1,"col2"]] - select multiple columns

df.loc[0] - select row by label

df.iloc[0] - select row by position

3. FILTERING DATA

df[df["col"]>100]

df[df["col2"]=="place"]

df[(df["col"]>100) & (df["col"]=="place")]

4. GROUPING AND AGGREGATION

df.groupby("col").sum() - df.groupby("city")["sales"].sum() # finds the sum

df.groupby("col").mean() - finds the avg

df.groupby("col").count() - counts the no of times

df.groupby("col").max() - finds the max 

df.groupby("col").min() - finds the min

5. DATA CLEANING

df.isnull() - shows missing values per col , it returns True if value is missing else False

df.isnull.sum() - counts how many missing values are in each col

df.dropna() - removes rows that contains missing values

df.fillna() - replaces missing values with smtg

df.drop_duplicates() - remove duplicate rows
