GROUP BY DATE RANGE

- means group data basedon time periods like :

- per day/week/month/blw 2 dates

EXAMPLE :

- jan total sales : 5000

- feb total sales : 7000

```python
import pandas as pd
data={
    "date":[
        "2024-01-01",
        "2024-01-05",
        "2024-02-10",
        "2024-02-15",
        "2024-03-20"
    ],
    "sales":[100,200,150,300,250]
}
df=pd.DataFrame(data)
df["date"]=pd.to_datetime(df["date"])
print(df)
```

```python
df.groupby(df["date"].dt.month)["sales"].sum()
```

```python
# grp by date range using resample()
# frst set data as index
#df.set_index("date",inplace=True) 
df.resample("ME").sum()
```

```python
# CUSTOM DATE RANGE GROUPING
df["period"]=pd.cut(
    df.index,
    bins=[                                      # bins : intervels
        pd.Timestamp("2024-01-01"),             # Timestamp : is a datetime object,which pandas understands for date comparision
        pd.Timestamp("2024-03-01"),
        pd.Timestamp("2024-05-01")
    ],
    labels=["jan-feb","mar-apr"]
)
df.groupby("period")["sales"].sum()
```

PRACTICE



```python
import pandas as pd
data={
    "date":[
        "2024-01-01",
        "2024-01-10",
        "2024-02-05",
        "2024-02-20",
        "2024-03-01"
    ],
    "sales":[80,90,85,70,95]
}
df=pd.DataFrame(data)
df["date"]=pd.to_datetime(df["date"])
```

```python
#1 group total sales by month
df.groupby(df["date"].dt.month)["sales"].sum()
```

```python
#2 group total sales by year
df.groupby(df["date"].dt.year)["sales"].sum()
```

```python
#3 find avg sales per month
df.groupby(df["date"].dt.month)["sales"].mean()
```

```python
#4 count no of sale records per month
df.groupby(df["date"].dt.month)["sales"].count()
```

```python
#5 find max sales in each month
df.groupby(df["date"].dt.month)["sales"].max()
```

```python
#6 set date as index and find monthly total sales
df.set_index("date",inplace=True) 
df.resample("ME").sum()
```

```python
#7 find weekly total sales
df.resample("W").sum()
```

```python
#8 find monthly avg sales
df.resample("ME").mean()
```

```python
#9 find no of sales per month
df.resample("ME").count()
```
