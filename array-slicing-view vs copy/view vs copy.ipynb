WHAT IS A VIEW ?

view = reference to original data

if uh change the view,the original array also changes bcoz view shares the original array

```python
import numpy as np
arr=np.array([10,20,30,40])
view_arr=arr[1:3]
view_arr[0]=99
print(view_arr)  
print(arr)    # changes in original also
```

WHAT IS COPY ?

- copy = completely separate data

- copy has its own memory

- if you change the copy , the original array does not change

```python
import numpy as np
arr=np.array([10,20,30,40])
copy_arr=arr[1:3].copy()
copy_arr[0]=99
print(arr)
print(copy_arr)    # changes only in copy
```
